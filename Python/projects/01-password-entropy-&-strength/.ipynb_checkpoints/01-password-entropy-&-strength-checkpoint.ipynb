# Lab 01: Análisis de Entropía y Robustez de Passwords 🛡️
**Objetivo:** Evaluar qué tan difícil es romper una contraseña mediante ataques de fuerza bruta usando lógica de Python.

```python
import re

def analizar_password(pwd):
    score = 0
    feedback = []
    
    # Criterios de longitud y complejidad
    if len(pwd) >= 12: score += 2
    elif len(pwd) >= 8: score += 1
    else: feedback.append("❌ Demasiado corta")

    if re.search(r"[A-Z]", pwd): score += 1
    else: feedback.append("❌ Falta mayúscula")

    if re.search(r"\d", pwd): score += 1
    else: feedback.append("❌ Falta número")

    if re.search(r"[!@#$%^&*]", pwd): score += 1
    else: feedback.append("❌ Falta caracter especial")

    return score, feedback

# PRUEBA TU CONTRASEÑA AQUÍ
mi_pass = "Argentina#1994" 
puntos, sugerencias = analizar_password(mi_pass)

print(f"Password: {mi_pass} | Seguridad: {puntos}/5")
for s in sugerencias: print(s)
```

```python
import matplotlib.pyplot as plt
import math

# Usamos log10 para que el número sea manejable (ej: 10^26 se convierte en 26)
labels = ['Números', 'Alfanumérica', 'Tu Password']
# Calculamos el logaritmo de las combinaciones para que matplotlib no explote
valores_log = [
    8, # 10^8
    math.log10(62**8), 
    len(mi_pass) * math.log10(94)
]

plt.figure(figsize=(9, 5))
plt.bar(labels, valores_log, color=['#e74c3c', '#f1c40f', '#2ecc71'])

plt.title('Fortaleza de la Contraseña (Escala Logarítmica)')
plt.ylabel('Potencia de 10 (Entropía)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
